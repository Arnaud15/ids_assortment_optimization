```python
import numpy as np
import seaborn as sns
sns.set_style('darkgrid')
```

```python
rewards = [np.random.randn(100) for _ in range(5)]

regrets = np.array([reward + 1. for reward in rewards]) 
```

```python
regrets.shape
```

```python
cumul = np.cumsum(regrets, axis=1)
```

```python
regrets
```

```python
cumul
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({"timestep":np.tile(np.arange(cumul.shape[1]), cumul.shape[0]), "signal":cumul.flatten()})
```

```python
sns.lineplot(data=df, x='timestep', y='signal', ci="sd")
```
